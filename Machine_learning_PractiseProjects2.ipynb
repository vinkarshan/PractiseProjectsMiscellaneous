##Waiter Tips graph creation
using ml model to predict waiter tips based upon past data. Data recorded - total bill, tip, gender of customer, customer smoker or not, day of week, lunch or dinner, size of people on table

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as graph_objects
data = pd.read_csv('/content/datasetmlpractise.csv')
print(data.head(100))
```

```python
print(data.isnull().sum())
```

```python
# billpaid-number of people at table- gender of person paying bill

figure = px.scatter(data_frame = data, x= 'total_bill',
                    y= 'tip', size = 'size', color = 'sex', trendline = 'ols')
figure.show()
```

```python
# bill paid- number of people at table- time of meal
figure = px.scatter(data_frame = data, x= 'total_bill',
                    y='tip', size = 'size', color = 'time', trendline = 'ols')
figure.show()
```

```python
# tip info according to day
figure = px.pie(data,
                values = 'tip',
                names = 'day', hole = 0.5)
figure.show()
```

```python
# tip receipt according to gender
figure = px.pie(data,
                values = 'tip',
                names = 'sex', hole = 0.5)
figure.show()
```

```python
# data transformation of categorical values into numerical values

data['sex'] = data['sex'].map({'Female':0, 'Male':1})
data['smoker'] = data['smoker'].map({'No': 0, 'Yes': 1})
data['day'] = data['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
data['time'] = data['time'].map({'lunch': 0, 'Dinner': 1})
data.head()
```
